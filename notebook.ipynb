```python
import pandas as pd

df_orders = pd.read_csv('data/olist_orders_dataset.csv')
df_customers = pd.read_csv('data/olist_customers_dataset.csv')
df_sellers = pd.read_csv('data/olist_sellers_dataset.csv')
df_items = pd.read_csv('data/olist_order_items_dataset.csv')
df_geolocations = pd.read_csv('data/olist_geolocation_dataset.csv')

print('Pedidos:')
display(df_orders.head())
print('Clientes:')
display(df_customers.head())
print('Vendedores:')
display(df_sellers.head())
print('Items:')
display(df_items.head())
print('Localizações:')
display(df_geolocations.head())
```

```python
cols_items = ['order_id','seller_id','price','freight_value','shipping_limit_date']
items_min = df_items[cols_items].copy()

# métricas por pedido
items_agg = (items_min
    .groupby('order_id')
    .agg(
        n_items=('seller_id','size'),
        n_sellers=('seller_id', pd.Series.nunique),
        price_sum=('price','sum'),
        freight_sum=('freight_value','sum'),
        ship_deadline_max=('shipping_limit_date','max'),
    )
    .reset_index()
)
```

```python
geo = (df_geolocations
       .sort_values(['geolocation_zip_code_prefix'])  # opcional
       .groupby('geolocation_zip_code_prefix', as_index=False)
       .agg(
           geo_lat=('geolocation_lat','median'),
           geo_lng=('geolocation_lng','median'),
           geo_city_mode=('geolocation_city', lambda x: x.mode().iat[0] if not x.mode().empty else x.iloc[0]),
           geo_state_mode=('geolocation_state', lambda x: x.mode().iat[0] if not x.mode().empty else x.iloc[0]),
       ))
```

```python
orders = df_orders[['order_id','customer_id','order_purchase_timestamp',
                    'order_delivered_customer_date','order_estimated_delivery_date','order_approved_at']].copy()

customers = df_customers[['customer_id','customer_zip_code_prefix','customer_city','customer_state']].copy()
sellers   = df_sellers[['seller_id','seller_zip_code_prefix','seller_city','seller_state']].copy()

# orders ⟵ customers  (m:1)
df = orders.merge(customers, on='customer_id', how='left', validate='m:1')

# orders ⟵ items_agg  (m:1)
df = df.merge(items_agg, on='order_id', how='left', validate='m:1')

items_sellers = (df_items[['order_id','seller_id']]
                 .merge(sellers, on='seller_id', how='left', validate='m:1'))
seller_feats = (items_sellers.groupby('order_id')
                .agg(n_seller_states=('seller_state', pd.Series.nunique))
                .reset_index())

df = df.merge(seller_feats, on='order_id', how='left', validate='m:1')

# Geoloc do cliente (m:1)
geo_cust = geo.rename(columns={'geolocation_zip_code_prefix':'customer_zip_code_prefix',
                               'geo_lat':'cust_lat','geo_lng':'cust_lng'})
df = df.merge(geo_cust[['customer_zip_code_prefix','cust_lat','cust_lng']],
              on='customer_zip_code_prefix', how='left', validate='m:1')

# Geoloc do seller: agregue por pedido (evita m:n)
items_seller_geo = (df_items[['order_id','seller_id']]
    .merge(sellers[['seller_id','seller_zip_code_prefix']], on='seller_id', how='left', validate='m:1')
    .merge(geo.rename(columns={'geolocation_zip_code_prefix':'seller_zip_code_prefix',
                               'geo_lat':'sell_lat','geo_lng':'sell_lng'})[['seller_zip_code_prefix','sell_lat','sell_lng']],
           on='seller_zip_code_prefix', how='left', validate='m:1'))

# tire média por pedido (se houver múltiplos sellers)
seller_geo_by_order = (items_seller_geo.groupby('order_id')
                       .agg(sell_lat_mean=('sell_lat','mean'),
                            sell_lng_mean=('sell_lng','mean'))
                       .reset_index())

df = df.merge(seller_geo_by_order, on='order_id', how='left', validate='m:1')

display(df.head())
```

```python
from geopy.distance import geodesic

def calculate_distance(row):
	if pd.isna(row['cust_lat']) or pd.isna(row['cust_lng']) or pd.isna(row['sell_lat_mean']) or pd.isna(row['sell_lng_mean']):
		return pd.NA
	return abs(geodesic((row['cust_lat'], row['cust_lng']), (row['sell_lat_mean'], row['sell_lng_mean'])).kilometers)

df['distance'] = df.apply(calculate_distance, axis=1)
df.head()
```

```python
num_cols = ['n_items','n_sellers','price_sum','freight_sum',
            'n_seller_states', 'processing_time','distance',
            'purchase_month', 'purchase_weekday','freight_ratio',
            'items_per_seller']

cat_cols = [c for c in X.columns if c not in num_cols]

def coerce_numeric(A):
    # A chega como array; reconstituímos como DataFrame para usar replace
    A = pd.DataFrame(A)
    A = A.replace({pd.NA: np.nan}).replace([np.inf, -np.inf], np.nan)
    return A.astype('float64').values
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import FunctionTransformer, OneHotEncoder

num_tf = Pipeline(steps=[
    ('tofloat', FunctionTransformer(coerce_numeric, feature_names_out='one-to-one')),
    ('imp', SimpleImputer(strategy='median')),
])

cat_tf = Pipeline(steps=[
    ('imp', SimpleImputer(strategy='most_frequent')),     # garante sem NaN
    ('ohe', OneHotEncoder(handle_unknown='ignore'))      
])


pre = ColumnTransformer(
    transformers=[
        ('num', num_tf, num_cols),
        ('cat', cat_tf, cat_cols),
    ],
    remainder='drop',
    verbose_feature_names_out=False
)

rf = Pipeline(steps=[
    ('pre', pre),
    ('model', RandomForestClassifier(
        n_estimators=400,
        max_depth=None,
        min_samples_split=2,
        min_samples_leaf=1,
        n_jobs=-1,
        class_weight='balanced',
        random_state=42
    ))
])

rf.fit(X_train, y_train)
```
